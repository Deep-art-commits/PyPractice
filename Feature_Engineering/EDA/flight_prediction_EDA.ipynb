## Starting with Feature Engineering 

```python
flight_df.head(1)
```

```python
# separating Datein to separate dates , months and year columns 
flight_df['Date']=flight_df['Date_of_Journey'].str.split("/").str[0].astype(int)
flight_df['Month']=flight_df['Date_of_Journey'].str.split("/").str[1].astype(int)
flight_df['Year']=flight_df['Date_of_Journey'].str.split("/").str[2].astype(int)
flight_df.head(1)
```

```python
flight_df.info()
#  month , date and year variables ae still oin object -string format , numeric conversion is needed 
```

```python
del flight_df['Date_of_Journey']
# or  we can use df.drop(variable , axis=1,inplace=True)
```

```python
flight_df.head(1)
```

```python
# splitting time from date in arrival time 
flight_df['Arrival_Time']=flight_df['Arrival_Time'].apply(lambda x:x.split(" ")[0])
```

```python
# Handling Arrival_Time
flight_df['arrival_hour']=flight_df['Arrival_Time'].str.split(":").str[0].astype(int)
flight_df['arrival_min']=flight_df['Arrival_Time'].str.split(":").str[1].astype(int)
flight_df.head(1)
```

```python
del flight_df['Arrival_Time']
```

```python
flight_df['departure_hour']=flight_df['Dep_Time'].str.split(":").str[0].astype(int)
flight_df['departure_min']=flight_df['Dep_Time'].str.split(":").str[1].astype(int)
flight_df.drop('Dep_Time',axis=1,inplace=True)
flight_df.head(1)
```

```python
flight_df['Total_Stops'].unique()
```

```python
flight_df['Total_Stops']=flight_df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, np.nan:1, '4 stops':4})
```

```python
flight_df.head(1)
```

```python
# handling duration 
flight_df['d_hr']=flight_df['Duration'].str.split(" ").str[0].str.split("h").str[0]
# handling duration 
flight_df['d_min']=flight_df['Duration'].str.split(" ").str[1].str.split("m").str[0]


```

```python
flight_df = flight_df[flight_df['d_hr'] != '5m']

# Convert duration columns to numeric and fill missing minutes with 0
flight_df['d_hr'] = flight_df['d_hr'].astype(int)
flight_df['d_min'] = flight_df['d_min'].fillna(0).astype(int)

# Calculate total duration in minutes
flight_df['duration_in_minutes'] = (flight_df['d_hr'] * 60) + flight_df['d_min']
flight_df.drop(['Duration','d_hr','d_min'],axis=1,inplace=True)
flight_df.head(2)
```

```python
flight_df.drop(['Route'],axis=1,inplace=True)
```

```python
flight_df.head()
```

```python
# checking unique values of all categorical variables
for col in flight_df.columns:
    if flight_df[col].dtype=='object':
        print(f"unique values for {col} are --->{flight_df[col].unique()}\n")
```

```python
from sklearn.preprocessing import OneHotEncoder
cat_encoder=OneHotEncoder()
```

```python
cat_columns=[col for col in flight_df.columns if flight_df[col].dtype=='object' ]
encodedcat_df=pd.DataFrame(cat_encoder.fit_transform(flight_df[cat_columns]).toarray(),columns=cat_encoder.get_feature_names_out())
```

```python
flight_transformed_df=pd.concat([flight_df,encodedcat_df],axis=1)
flight_transformed_df.drop(cat_columns,axis=1,inplace=True)
print(flight_transformed_df.shape)
flight_transformed_df.head()
```
