```python
import pandas as pd
import numpy as np 
import random as r
np.random.seed(123)
n_samples=1000
n_ratio_c1=0.9
n_samples_class1=int(n_samples*n_ratio_c1)
n_samples_class2=n_samples-n_samples_class1
print(n_samples_class1,n_samples_class2)
```

```python
# creating the imbalance dataset
class0=pd.DataFrame(
    {
        'feature 1':np.random.normal(loc=0,scale=1,size=n_samples_class1),
        'feature 2':np.random.normal(loc=0,scale=1,size=n_samples_class1),
        'target':[0]*n_samples_class1
        
    }
)
class1=pd.DataFrame(
    {
        'feature 1':np.random.normal(loc=2,scale=1,size=n_samples_class2),
        'feature 2':np.random.normal(loc=2,scale=1,size=n_samples_class2),
        'target':[1]*n_samples_class2
        
    }
)
```

```python
df=pd.concat([class0,class1]).reset_index(drop=True)
df.head()
```

```python
df['target'].value_counts()
```

```python
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
from sklearn.utils import resample
upsampled_min_df=resample(df_minority,replace=True,\
                          n_samples=len(df_majority),random_state=42  )
upsampled_min_df.shape
```

```python
df_upsampled=pd.concat([df_majority,upsampled_min_df])
df_upsampled['target'].value_counts()
```

```python
downsampled_maj_df=resample(df_majority,n_samples=len(df_minority),random_state=43)
downsampled_maj_df.shape
```

```python
df_downsampled=pd.concat([df_minority,downsampled_maj_df])
df_downsampled['target'].value_counts()
```
