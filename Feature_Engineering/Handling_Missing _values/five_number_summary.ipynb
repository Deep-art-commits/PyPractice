### 
five number summary stands for 
### min , max , Quartile1, q3, median and IQR(q3-q1)

```python
import random as r
lst_marks=[r.randint(1,100) for i in range(1,20)]
# lst_marks
```

```python
# calculating 5 point summary 
import numpy as  np
min,q1,median,q3,max=np.quantile(lst_marks,[0.0,0.25,0.5,0.75,1.0])
```

```python
IQR=q3-q1
IQR
```

```python
lower_fence=q1-1.5*(IQR)
upper_fence=q3+1.5*(IQR)
```
