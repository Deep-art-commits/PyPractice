#### Label encoder assigns distinct numerical labels to categorical values either based on value or frequency 


```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
l_enc=LabelEncoder()
```

```python
df=pd.DataFrame(
    {
        'color':['red','blue','green','green','red','blue']
    }
)
df
```

```python
label_encoded=l_enc.fit_transform(df[['color']]).toarray()
```
