```python
import matplotlib.pyplot as plt
```

```python
x=[i for i in range(1,6)]
y=[i**3 for i in range(7,12)]
plt.plot(x,y,color='red',linestyle="--",marker='o',linewidth=5,markersize=9)

plt.xlabel("x_axis")
plt.ylabel("y_axis") 

# formating the chart 
plt.show()
```

```python
# Multiple Plots
import random as r

x=[ r.randint(1,20) for i in range(1,10)]
y=[r.randint(20,40) for i in range(1,10)]
z=[r.randint(80,100) for i in range(1,10)]

plt.Figure(figsize=(9,5))
plt.subplot(2,2,1)
plt.plot(x,y,color='r')
plt.title("Plot 1")


plt.subplot(2,2,2)
plt.plot(x,z,color='g')
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(z,y,color='b')
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(z,x,color='g')
plt.title("Plot 4")
```

```python
# Bar plot
categies=['A','B','C','D']
values=[20,59,52,60]

plt.bar(categies,values,color='purple')
```

```python
# hostogram
data=[r.randint(1,1000) for i in range(1,20)]
plt.hist(data,bins=3,color='orange',edgecolor='black')
```
