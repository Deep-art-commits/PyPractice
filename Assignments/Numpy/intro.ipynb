### Numpy  is a package used to perform mathematicval operations

#### np.array is used to create a numpy array

```python
import numpy as np
arr1=np.array([1,2,3,4])
print(arr1)
# printinng shape of an array
print(arr1.shape)
print(type(arr1))
arr3=np.array([[1,2,3,4],[4,5,6,7]])
print(arr3)
print(arr3.shape)


# using np.arange to create a nimpy array 
arr4=np.arange(0,10,2).reshape(5,1)
print(arr4)
print(arr4.shape)
```

### Reshaping an array
#### Can only be reshaped in multiples , in case of 1D array , the number of  total elements should match 

```python
arr2=arr1.reshape(2,2)
arr2
```

#### Numpy vectorised operations
##### Addition/Substraction/Multipication/Division

```python
a1=np.arange(1,6)
a2=np.arange(10,15)
print(f"a1={a1}\na2={a2}")
print("*****************************************")
print(f"Addition --> {a1+a2}")
print("*****************************************")
print(f"Substraction --> {a1-a2}")
print("*****************************************")
print(f"Multipication --> {a1*a2}")
print("*****************************************")
print(f"Division --> {a1/a2}")
```

Universal functions


```python
# sqrt
arr=[2,3,4,5,6]
print(np.sqrt(arr))
print(np.exp(arr))
print(np.sin(arr))
print(np.log(arr))
```
