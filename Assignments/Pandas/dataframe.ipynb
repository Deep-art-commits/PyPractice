### Creating Dataframe

```python
import pandas as pd
```

```python
# using a dictionary of lists

data={
    'name':['Dilpreet','Anmol','Andeep'],
    'age':[34,32,2],
    'city':['Bla','sgr','Bla']
    }
df=pd.DataFrame(data)
df
```

### Accessing Data elements
