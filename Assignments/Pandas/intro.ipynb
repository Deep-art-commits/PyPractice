### Pandas is a powerful data manipulation library widely used for data manipulation and cleaning
#### Provides 2 Datastructures----->Series and Dataframe
## Series 
### is a 1D array like object 
## Dataframe
### is 2D structure, size mutable , tabular heterogenous structure  with labelled axes (rows and columns)

```python
import pandas as pd
```

##### Creating series 


```python
# using list
a=[1,2,3,4,5]
ser=pd.Series(a)
print(f"series-->\n{ser}-->{type(ser)}")
```

```python
# using dictionary
dict={'a':1,"b":2,'c':3}
s_d=pd.Series(dict)
print(f"series-->\n{s_d}-->{type(s_d)}")
# keys become the indec in this case 
```

```python
l=[1,2,3]
id=['a','b','c']
s_d1=pd.Series(l,index=id)
print(f"series-->\n{s_d1}-->{type(s_d1)}")
```
