```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.scatterplot(tips,x='total_bill',y='tip')
```

```python
sns.lineplot(x='size',y='tip',data=tips)
```

```python
sns.histplot(data=tips['total_bill'],bins=10,kde=True)
```
