Create a list of random numbers and sort it in ascending and descending order. Remove the duplicates from the list and print the modified list.

```python
import random   
ran_list=[random.randint(1, 100) for _ in range(20)]
print(f"Random list of 20 integers: {ran_list}")
asc_sorted=sorted(ran_list)
print(f"Ascending sorted list: {asc_sorted}")
desc_sorted=sorted(ran_list, reverse=True)
print(f"Descending sorted list: {desc_sorted}")
dup_removed=list(set(ran_list))
print(f"List with duplicates removed: {dup_removed}")
```
